# 04. Answers - Exercise 4.8

```python
# group and aggregate for the entire ords_prods_merge dataframe
ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

- In the subset only 8 department_id's are listed but all all 21 are listed when the whole dataframe is grouped and aggregated.
- This means that the highest order_number in the dataframe is higher than in the subset.

Question 4

```python
# create column 'max_order' to retrieve max value for order_number
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
# assigning loyalty status for loyal customers
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40 , 'loyalty_flag'] = 'Loyal Customer'
```

```python
# assigning loyalty status for regular customers
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
# assigning loyalty status for new customers
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

Question 5

```python
# shows mean, min and max for prices paid by grouping 
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

Question 6

```python
# create average_price column and calculate average price grouped by user_id
ords_prods_merge['average_price'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
# create a spending_flag and provide label for 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# create a spending_flag and provide label for 'High Spender'
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

Question 7

```python
ords_prods_merge['order_days_median'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.loc[ords_prods_merge['order_days_median'] > 20, 'order_frequency_flag'] = 'Non-Frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_days_median'] > 10) & (ords_prods_merge['order_days_median'] <= 20), 'order_frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_days_median'] <= 10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# counting the actual number of uniques users and their frequencies in the column
ords_prods_merge.groupby('order_frequency_flag')['user_id'].nunique()
```

```python
# creating a subset to investigate NaN values in frquencies
df_3 = ords_prods_merge[ords_prods_merge['order_frequency_flag'].isnull()]
```

```python
# retrieve the rows where cutomers are not classified (NaN)
df_3
```

- The NaN values are due to the missing values in the 'days_since_prior_order' column
- They are all from one customer, for one specific order

```python
# replacing NaN values with Non-Frequent Customer
ords_prods_merge['order_frequency_flag'].fillna('Non-Frequent Customer', inplace = True)
```

```python
# confirming that NaN values have been filled in
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
